## Introduction to Regularization

**Regularization** is a technique used in regression models to prevent overfitting. Overfitting occurs when a model learns not only the underlying patterns in the training data but also the noise, leading to poor performance on new, unseen data. Regularization adds a penalty term to the loss function used to train the model, discouraging the model from fitting the noise in the training data.

Two common types of regularization used in regression are **Ridge Regression** and **Lasso Regression**. Both techniques aim to improve the generalizability of the model by adding a penalty for large coefficients, but they do so in different ways.

## Explanation of Ridge Regression

**Ridge Regression** (also known as Tikhonov regularization) adds a penalty equal to the sum of the squared values of the coefficients (also known as the L2 penalty) to the loss function. The Ridge Regression objective function is:

$$ \text{Ridge Loss} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} \beta_j^2 $$

Where:
- $( y_i $ ): Actual value
- $( \hat{y}_i $ ): Predicted value
- $( \beta_j $ ): Coefficients
- $( \lambda $ ): Regularization parameter (controls the strength of the penalty)

## Explanation of Lasso Regression

**Lasso Regression** (Least Absolute Shrinkage and Selection Operator) adds a penalty equal to the sum of the absolute values of the coefficients (also known as the L1 penalty) to the loss function. The Lasso Regression objective function is:

$$ \text{Lasso Loss} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} |\beta_j| $$

Where:
- $( y_i $): Actual value
- $( \hat{y}_i $): Predicted value
- $( \beta_j $): Coefficients
- $( \lambda $): Regularization parameter (controls the strength of the penalty)

## Benefits of Ridge and Lasso Regression

- **Prevent Overfitting**: Both methods add a penalty for large coefficients, discouraging the model from fitting the noise in the training data.
- **Model Simplicity**: Lasso Regression can shrink some coefficients to exactly zero, effectively performing feature selection and resulting in simpler models.
- **Improved Generalization**: By preventing overfitting, these techniques improve the model's ability to generalize to new, unseen data.

<hr>

## Differences Between Ridge and Lasso Regression

- **Penalty Type**: Ridge uses **L2 penalty** (squared coefficients), while Lasso uses **L1 penalty** (absolute coefficients).
- **Feature Selection**: *Lasso can shrink coefficients to zero*, effectively selecting a subset of features, while *Ridge tends to shrink coefficients but not eliminate them entirely*.
- **Model Interpretability**: Lasso can produce more interpretable models with fewer features.

## Methods for Implementing Ridge and Lasso Regression

Ridge and Lasso Regression can be implemented using programming libraries such as `sklearn` in Python. Below, we'll focus on implementing these techniques using `sklearn`.


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Generating some random data for demonstration
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
Y = 4 + 3 * X + np.random.randn(100, 1)
```

## Implementing **Ridge Regression** using `sklearn`


```python
# Creating a Ridge Regression model using sklearn
ridge_model = Ridge(alpha=1.0)

# Fitting the model
ridge_model.fit(X, Y)

# Predicting
Y_pred_ridge = ridge_model.predict(X)

# Evaluating the model
mse_ridge = mean_squared_error(Y, Y_pred_ridge)
r2_ridge = r2_score(Y, Y_pred_ridge)

print(f"Mean Squared Error (Ridge): {mse_ridge}")
print(f"R-squared (Ridge): {r2_ridge}")
```

## Implementing **Lasso Regression** using `sklearn`


```python
# Creating a Lasso Regression model using sklearn
lasso_model = Lasso(alpha=0.1)

# Fitting the model
lasso_model.fit(X, Y)

# Predicting
Y_pred_lasso = lasso_model.predict(X)

# Evaluating the model
mse_lasso = mean_squared_error(Y, Y_pred_lasso)
r2_lasso = r2_score(Y, Y_pred_lasso)

print(f"Mean Squared Error (Lasso): {mse_lasso}")
print(f"R-squared (Lasso): {r2_lasso}")
```

```python
# Plotting the results
plt.figure(figsize=(12, 6))

# Plotting X vs Y
plt.scatter(X, Y, color='blue', label='Actual data')
plt.plot(X, Y_pred_ridge, color='red', label='Ridge Regression')
plt.plot(X, Y_pred_lasso, color='green', label='Lasso Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Ridge and Lasso Regression')
plt.legend()
plt.show()
```
