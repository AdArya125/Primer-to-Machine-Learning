#### Explanation of Bayesian Regression and Its Applications

**Bayesian Regression** is a statistical technique that applies Bayesian principles to linear regression. In Bayesian regression, we incorporate prior beliefs about the parameters and update these beliefs as we observe data. This results in a posterior distribution for the parameters, which represents our updated beliefs after seeing the data.

**Key Concepts**:
- **Prior Distribution**: Represents our initial beliefs about the parameters before observing any data.
- **Likelihood**: The probability of the observed data given a set of parameters.
- **Posterior Distribution**: The updated belief about the parameters after observing the data, obtained by combining the prior distribution and the likelihood using Bayes' theorem.

**Applications of Bayesian Regression**:
- **Uncertainty Quantification**: Provides a probabilistic framework to quantify uncertainty in parameter estimates and predictions.
- **Regularization**: Naturally incorporates regularization by specifying priors, which can prevent overfitting.
- **Model Comparison**: Facilitates model comparison and selection by computing the marginal likelihood of different models.

## Bayesian Regression Formulation

In Bayesian regression, we aim to find the posterior distribution of the parameters given the data. The key components are:

1. **Prior Distribution**:
   The prior distribution represents our beliefs about the parameters before seeing the data. For example, we might assume a normal prior for the parameters:
   $$\beta \sim \mathcal{N}(\mu_\beta, \sigma_\beta^2) $$

2. **Likelihood**:
   The likelihood represents the probability of the observed data given the parameters. Assuming normally distributed errors, the likelihood can be written as:
   $$y \mid X, \beta, \sigma^2 \sim \mathcal{N}(X\beta, \sigma^2)$$
3. **Posterior Distribution**:
   Using Bayes' theorem, we combine the prior and the likelihood to obtain the posterior distribution of the parameters:
   $$p(\beta \mid X, y) \propto p(y \mid X, \beta) \cdot p(\beta)$$

The posterior distribution provides a complete description of our updated beliefs about the parameters after observing the data.


## Bayesian Inference Steps

1. **Specify the Model**:
   Define the prior distributions for the parameters and the likelihood function for the observed data.

2. **Collect Data**:
   Gather the observed data that will be used to update our beliefs.

3. **Compute the Posterior**:
   Use Bayes' theorem to combine the prior distribution and the likelihood to obtain the posterior distribution.

4. **Make Predictions**:
   Use the posterior distribution to make predictions and quantify uncertainty.

5. **Model Evaluation**:
   Evaluate the model using appropriate metrics and compare different models using criteria like the marginal likelihood or information criteria (e.g., AIC, BIC).


```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import BayesianRidge
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Generating some random data for demonstration
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
Y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
# Creating a Bayesian Ridge Regression model
bayesian_ridge_model = BayesianRidge()

# Reshape Y to a 1D array
Y = Y.ravel()

# Fitting the model
bayesian_ridge_model.fit(X, Y)

# Predicting
Y_pred_bayesian = bayesian_ridge_model.predict(X)

# Evaluating the model
mse_bayesian = mean_squared_error(Y, Y_pred_bayesian)
r2_bayesian = r2_score(Y, Y_pred_bayesian)

print(f"Mean Squared Error (Bayesian Ridge): {mse_bayesian}")
print(f"R-squared (Bayesian Ridge): {r2_bayesian}")
```

```python
# Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Actual data')
plt.plot(X, Y_pred_bayesian, color='red', linewidth=3, label='Bayesian Ridge Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bayesian Ridge Regression')
plt.legend()
plt.show()
```
