# 9.2.1 Moving Average

## Explanation of Moving Average

A Moving Average (MA) is a statistical technique used to analyze time series data by smoothing out short-term fluctuations and highlighting longer-term trends or cycles. The primary idea behind a moving average is to calculate the average of data points within a specific window or period and then move this window across the dataset to create a smoothed series.

There are several types of moving averages, including:

- **Simple Moving Average (SMA)**: The average of data points within a fixed window, calculated by summing up the values and dividing by the number of periods.
- **Weighted Moving Average (WMA)**: Similar to SMA but assigns different weights to different data points within the window, giving more importance to recent values.
- **Exponential Moving Average (EMA)**: A type of weighted moving average that gives more weight to recent observations, allowing it to respond more quickly to new data.

### Mathematical Formula

For a Simple Moving Average (SMA) with window size $ w $:

$$ \text{SMA}_t = \frac{1}{w} \sum_{i=0}^{w-1} x_{t-i} $$

where:
- $ \text{SMA}_t $ is the moving average at time $t $.
- \( x_{t-i} \) represents the data points in the window.

## Benefits and Use Cases of Moving Average

### Benefits
1. **Trend Identification**: Moving averages help in identifying underlying trends by smoothing out noise and short-term fluctuations.
2. **Signal Filtering**: They filter out noise, making it easier to see the genuine trends and patterns in the data.
3. **Forecasting**: Moving averages are often used in forecasting models to predict future values based on historical trends.

### Use Cases
1. **Financial Markets**: Moving averages are widely used in stock market analysis to identify trends and signals for trading strategies.
2. **Economic Indicators**: They are used to smooth out economic data, such as GDP growth rates or unemployment rates, to better understand economic conditions.
3. **Sales and Marketing**: Businesses use moving averages to analyze sales trends, customer demand, and inventory levels, helping in inventory management and strategic planning.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.datasets import get_rdataset
```

```python
# Load the Air Passengers dataset
data = get_rdataset('AirPassengers', 'datasets')
df = data.data
```

```python
# Convert fractional years to datetime
# Assuming the data starts from January 1949 and each entry is a month apart
start_date = '1949-01-01'
df['Date'] = pd.date_range(start=start_date, periods=len(df), freq='M')
df.set_index('Date', inplace=True)
```

```python
# Compute Simple Moving Average
window_size = 12
df['SMA'] = df['value'].rolling(window=window_size).mean()
```

```python
# Plotting
plt.figure(figsize=(14, 12))

# Plot Original Data
plt.subplot(3, 1, 1)
plt.plot(df.index, df['value'], color='blue')
plt.title('Original Air Passengers Data')
plt.ylabel('Number of Passengers')
plt.grid(True)

# Plot Simple Moving Average
plt.subplot(3, 1, 2)
plt.plot(df.index, df['SMA'], color='red')
plt.title(f'Simple Moving Average (window={window_size})')
plt.ylabel('SMA')
plt.grid(True)

# Plot Original Data and SMA
plt.subplot(3, 1, 3)
plt.plot(df.index, df['value'], label='Original Data', color='blue')
plt.plot(df.index, df['SMA'], label=f'SMA (window={window_size})', color='red')
plt.title('Original Data with SMA')
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
