```python
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, uniform, expon, binom, poisson, geom
```

```python
# Setting Seaborn style for plots
sns.set(style="whitegrid")

# Function to plot probability distributions
def plot_distribution(distribution, title, continuous=True):
    plt.figure(figsize=(10, 6))
    if continuous:
        plt.plot(distribution, linewidth=2)
    else:
        plt.bar(np.arange(len(distribution)), distribution, color='skyblue')
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.show()
```

## 1. Continuous Probability Distributions:

- # Normal Distribution
<br>



```python
# Normal Distribution
mu, sigma = 0, 1  # mean and standard deviation
x = np.linspace(-4, 4, 100)
normal_distribution = norm.pdf(x, mu, sigma)
plot_distribution(normal_distribution, 'Normal Distribution')
```

- # Uniform Distribution
<br>




```python
# Uniform Distribution
uniform_distribution = uniform.pdf(x, loc=-2, scale=4)  # loc = lower bound, scale = width (b-a)
plot_distribution(uniform_distribution, 'Uniform Distribution')
```

- # Exponential Distribution
<br>



```python
# Exponential Distribution
exponential_distribution = expon.pdf(x, loc=0, scale=1)  # loc = starting point, scale = rate
plot_distribution(exponential_distribution, 'Exponential Distribution')
```

## 2. Discrete Probability Distributions:

- # Binomial Distribution
  <br>
  
  

```python
# Binomial Distribution
n, p = 10, 0.5  # number of trials, probability of success
k = np.arange(0, 11)
binomial_distribution = binom.pmf(k, n, p)
plot_distribution(binomial_distribution, 'Binomial Distribution', continuous=False)
```

- # Poisson Distribution
<br>



```python
# Poisson Distribution
lambda_ = 3  # average rate of occurrence
k = np.arange(0, 10)
poisson_distribution = poisson.pmf(k, lambda_)
plot_distribution(poisson_distribution, 'Poisson Distribution', continuous=False)
```

- # Geometric Distribution
<br>



```python
# Geometric Distribution
p = 0.3  # probability of success
k = np.arange(1, 11)
geometric_distribution = geom.pmf(k, p)
plot_distribution(geometric_distribution, 'Geometric Distribution', continuous=False)
```
